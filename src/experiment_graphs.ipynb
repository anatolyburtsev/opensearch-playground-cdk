```python
import matplotlib.pyplot as plt

# Data
docs_to_search = [100, 1000, 3000, 5000, 10000]
multiple_search_times = [1.18, 3.41, 9.37, 6.17, 15.59]
num_threads = [100, 50, 25, 12]
time_threads = [
    [1.73, 0.71, 0.83, 1.03],
    [3.49, 5.03, 6.59, 14.87],
    [14.06, 15.24, 19.37, 30.30],
    [22.03, 22.62, 28.31, 50.56],
    [40.36, 57.31, 64.63, 105.58]
]

# Create a figure with 5 subplots
fig, axs = plt.subplots(5, 1, figsize=(8, 40))

for i, ax in enumerate(axs):
    # Plot the time_threads data
    ax.plot(num_threads, time_threads[i], label='num_threads')

    # Plot a horizontal line for the multiple search time
    ax.hlines(y=multiple_search_times[i], xmin=12, xmax=100, colors='red', linestyles='dashed', label='multiple_search')

    # Setting labels and title
    ax.set_xlabel("Number of Threads")
    ax.set_ylabel("Time (seconds)")
    ax.set_title(f"Time to Search {docs_to_search[i]} Documents")

    # Show the legend
    ax.legend()

# Adjust spacing
plt.tight_layout()

# Show the plots
plt.show()
```
